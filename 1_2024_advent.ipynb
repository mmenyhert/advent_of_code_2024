```python
left = []
right = []
with open('input.txt') as f:
    lines = f.readlines()
    for line in lines:
        data = line.split('   ')
        left.append(int(data[0]))
        right.append(int(data[1]))
```

```python
sum([abs(l-r) for l, r in zip(sorted(left), sorted(right))])
```

```python
from collections import Counter
right_counter = Counter(right)
sum([(l * right_counter.get(l, 0)) for l, r in zip(sorted(left), sorted(right))])
```
