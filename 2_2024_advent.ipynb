```python
levels = [list(map(int, l.split(' '))) for l in open('input.txt')]

def increasing(a):
    return all((x < y for x, y in zip(a[:-1], a[1:])))

def decreasing(a):
    return increasing(a[::-1])

def max_diff(a, low = 1, high= 3):
    return all(( low <= abs(x-y) <= high for x, y in zip(a[:-1], a[1:]) ))

def safe(a, low= 1, high = 3):
    return (increasing(a) or decreasing(a)) and max_diff(a, low, high)

sum(list(map(safe, levels)))
```

```python
def remove_idx(a, idx):
    return a[:idx] + a[idx+1 :]

def safe_damp(a, low=1, high=3):
    return any(
        safe(remove_idx(a, idx), low, high) for idx in range(len(a))
    )


sum(list(map(safe_damp, levels)))
```
