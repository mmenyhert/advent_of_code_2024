```python
rule_pairs = [
    tuple(map(int,line.strip().split('|'))) for line in open('input.txt') if '|' in line
]

updates = [
    list(map(int, line.strip().split(','))) for line in open('input.txt') if ',' in line
]
```

```python
rules = {}
for rule_pair in rule_pairs:
    x, y = rule_pair
    rules[y] = rules.get(y, []) + [x]

# rules[k] -> [denying numbers]
```

```python
def is_valid_update(update, rules):
    for i, e in enumerate(update[:-1]):
        rest = update[(i+1):]
        must_not_be_after = rules[e]
        for f in rest:
            if f not in must_not_be_after:
                pass
            else:
                return False
    return True

valid_updates = [update for update in updates if is_valid_update(update, rules)]
```

```python
sum(
    u[len(u)//2] for u in valid_updates
)
```

```python
invalid_updates = [update for update in updates if not is_valid_update(update, rules)]
```

```python
from functools import cmp_to_key
def is_correct_order(a, b):
    return b not in rules[a]

def compare(a, b):
    return int(is_correct_order(a,b)) * 2 -1


invalid_updates_sorted = [sorted(invalid_update, key=cmp_to_key(compare)) for invalid_update in invalid_updates]
```

```python
sum(
    u[len(u)//2] for u in invalid_updates_sorted
)
```
